```python
import investpy as inv
from datetime import date

data_atual = date.today()
data = '{}/{}/{}'.format(data_atual.day, data_atual.month,data_atual.year)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib
matplotlib.rcParams['figure.figsize'] = (16,8)
```

## Commodities



Cobre

```python
copper_results = inv.search_quotes(text = 'copper', products=['commodities'], countries=['united states'], n_results=50)
for copper_results in copper_results:
  print(copper_results)
```

```python
copper = copper_results.retrieve_historical_data(from_date='01/01/2019', to_date= data)
copper
```
